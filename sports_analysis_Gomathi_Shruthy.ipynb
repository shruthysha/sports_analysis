```python
def extract_specific_fields(person_detail_string):
    # Split the string into a list of fields
    fields = person_detail_string.split(',')

    # Extract the desired fields (name, age, batting style, bowling style)
    name = fields[0]
    age = fields[1]
    batting_style = fields[2]
    bowling_style = fields[3]

    # Return a dictionary containing the extracted fields
    return {
        'Name': name,
        'Age': age,
        'Batting Style': batting_style,
        'Bowling Style': bowling_style
    }

# Apply the function to the 'PersonDetail' column and create a new DataFrame
df_new = df['PersonDetail'].apply(extract_specific_fields).to_frame().T

# Display the new DataFrame
print(df_new)
```
